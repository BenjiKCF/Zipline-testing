```python
import pandas as pd
from collections import OrderedDict
import pytz
from zipline.api import order, record, symbol, set_benchmark
import zipline
from datetime import datetime
```

```python
# order dict to contain all dataframe > convert to pandas df
data = OrderedDict()
data['SPY'] = pd.read_csv("SPY.csv", index_col = 0, parse_dates=['date'])
data['SPY'] = data['SPY'][['open', 'high', 'low', 'close', 'volume']]
```

```python
data['SPY'].head()
```

```python
panel = pd.Panel(data)
panel.minor_axis = ['open', 'high', 'low', 'close', 'volume']
panel.major_axis = panel.major_axis.tz_localize(pytz.utc)
print(panel)
```

```python
def initialize(context):
    set_benchmark(symbol("SPY"))


def handle_data(context, data):
    order(symbol("SPY"), 10)
    record(SPY=data.current(symbol('SPY'), 'price'))
    
perf = zipline.run_algorithm(start=datetime(2017, 1, 5, 0, 0, 0, 0, pytz.utc),
                      end=datetime(2018, 3, 1, 0, 0, 0, 0, pytz.utc),
                      initialize=initialize,
                      capital_base=100000,
                      handle_data=handle_data,
                      data=panel)
```

```python
perf.head()
```

```python
import matplotlib.pyplot  as plt

perf.portfolio_value.plot()
plt.show()
```
